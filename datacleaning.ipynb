```python
# Reclassify DIABETE4 to match the new ordinal values:
# 0: No diabetes or only during pregnancy
# 1: Pre-diabetes or borderline diabetes
# 2: Yes diabetes

# Replace the values in DIABETE4 column
brfss_df_selected['DIABETE4'] = brfss_df_selected['DIABETE4'].replace({
    1: 2,  # Yes
    2: 0,  # Yes, but female told only during pregnancy
    3: 0,  # No
    4: 1,  # Pre-diabetes or borderline diabetes
})

# Remove rows with 'Don't know' (7) and 'Refused' (9)
brfss_df_selected = brfss_df_selected[~brfss_df_selected['DIABETE4'].isin([7, 9])]

# Verify the unique values in DIABETE4
print(brfss_df_selected['DIABETE4'].unique())
```

##  Creating a Binary Dataset for Diabetic vs. Non-Diabetic


```python
# Make a copy of the cleaned dataframe
brfss_binary = brfss.copy()

# Reclassify the pre-diabetics instances from 1 to 0 and the diabetics instances from 2 to 1, so that 0 meaning non-diabetic and pre-diabetic and 1 meaning diabetic.
brfss_binary['Diabetes_012'] = brfss_binary['Diabetes_012'].replace({1:0})
brfss_binary['Diabetes_012'] = brfss_binary['Diabetes_012'].replace({2:1})

# Change the column name to Diabetes_binary
brfss_binary = brfss_binary.rename(columns = {'Diabetes_012': 'Diabetes_binary'})
brfss_binary.Diabetes_binary.unique()
```

```python
# Confirm the change on the column name
brfss_binary.head()
```

```python
# Confirm the class reclassification was done successfully
brfss_binary.groupby(['Diabetes_binary']).size()
```

```python
# Separate the 0 (Non-diabetes & Pre-diabetes) and 1 (Diabetes)
# Class 1 (diabetes group) dataframe
brfss_5050_1 = brfss_binary[brfss_binary['Diabetes_binary'] == 1]

# Class 0 (non-diabetes & pre-diabetes group) dataframe
brfss_5050_0 = brfss_binary[brfss_binary['Diabetes_binary'] == 0]

# Randomly select the same number of cases as the diabetes group from no-diabetes group (to balance classes)
num_diabetes_cases = len(brfss_5050_1)
brfss_5050_0_rand = brfss_5050_0.sample(n=num_diabetes_cases, random_state=42)

# Combine the two groups into one balanced DataFrame
brfss_5050 = pd.concat([brfss_5050_0_rand, brfss_5050_1], ignore_index=True)

# Verify the new balanced dataset
print(brfss_5050['Diabetes_binary'].value_counts())
```

```python
# Check the rows and columns of two dataframes: brfss_5050 and brfss_binary 
print(f'brfss_5050={brfss_5050.shape}',f'brfss_binary={brfss_binary.shape}')
```

```python
#Save the 50-50 balanced dataset to csv
#************************************************************************************************
brfss_5050.to_csv('Data/diabetes_binary_5050split_health_indicators_BRFSS2023.csv', sep=",", index=False)
#************************************************************************************************

#Also save the original binary dataset to csv
#************************************************************************************************
brfss_binary.to_csv('Data/diabetes_binary_health_indicators_BRFSS2023.csv', sep=",", index=False)
#************************************************************************************************
```
